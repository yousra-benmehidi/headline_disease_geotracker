**Project Workflow:**

1.  Load the data.
2.  Extract locations from the text using regular expressions and the GeoNames Cache library.
3.  Check the location matches for errors.
4.  Cluster the locations based on geographic distance.
5.  Visualize the clusters on a map, and remove any errors.
6.  Output representative locations from the largest clusters.
7.  Draw interesting conclusions.

```python
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.cluster import DBSCAN
from cartopy.crs import PlateCarree
import cartopy
from geonamescache import GeonamesCache
from unidecode import unidecode
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
```

**1. Load the data**

```python
#load the data
headlines_object=open("headlines.txt","r")
headlines=[line.strip() for line in headlines_object.readlines()]

print(len(headlines))
```

```python
#create world and duplicates dictionaries
gc = GeonamesCache()
countries = gc.get_countries_by_names()
cities = gc.get_cities()

world={}
for city in cities.values():
    for country_name,country_info in countries.items():
        if country_info['iso']==city['countrycode']:
            country_dic={
                'countryname':country_name,
                'country_pattern': re.compile(rf"\b{country_name}\b"),
            }
            break
            
    world[f"{city['name']}_{city['countrycode']}"]={
        'cityname':city['name'],
        'nonacc_cityname':unidecode(city['name']),
        'city_pattern': re.compile(rf"\b{unidecode(city['name'])}\b"),
        'latitude': city['latitude'],
        'longitude': city['longitude'],
        'country': country_dic
    }

duplicates={}
for city in cities.values():
    duplicated_cities=gc.get_cities_by_name(city['name'])
    if len(duplicated_cities)>1 and f"{city['name']}_{city['countrycode']}" not in duplicates:
        for dupcity in duplicated_cities:
            dupcity_info=list(dupcity.values())[0]
            duplicates[f"{dupcity_info['name']}_{dupcity_info['countrycode']}"]={
                'cityname':dupcity_info['name'],
                'countrycode':dupcity_info['countrycode'],
                'population': dupcity_info['population']
            }
```

**2. Extract the locations from text**

```python
#extract locations from headlines
locations={}
for headline in headlines:
    for city_id, city in world.items():
        match=city['city_pattern'].search(headline)
        if match and city_id not in locations:
            if city_id not in duplicates:
                locations[city_id]={
                    'latitude':city['latitude'],
                    'longitude':city['longitude'],
                }
                break
            else:
                for key,value in duplicates.items():
                    if value['cityname']==city['cityname'] and world[key]['country']['country_pattern'].search(headline):
                        locations[key]={
                            'latitude':world[key]['latitude'],
                            'longitude':world[key]['longitude'],
                        }
                        break
                #in case no country is found (only city is mentionned in headline and it's a duplicated name)-> pick most populated
                # -> pick the most populated duplicate
                same_city_duplicates = {
                    dup_key: dup_val for dup_key, dup_val in duplicates.items()
                    if dup_val['cityname'] == city['cityname']
                }
            
                if same_city_duplicates:
                    # Pick the duplicate entry with the largest population
                    max_key = max(same_city_duplicates, key=lambda k: same_city_duplicates[k]['population'])
                    locations[max_key] = {
                        'latitude': world[max_key]['latitude'],
                        'longitude': world[max_key]['longitude'],
                    }
                    break
                
```

**3. Check the location matches for errors.**

```python
#turn locations into datapoints
dp=[]
for key,value in locations.items():
    lat=value['latitude']
    lon=value['longitude']
    if -90<=lat<=90 and -180<=lon<=180:
        dp.append((value['latitude'],value['longitude']))
```

```python
plt.figure(figsize=(12, 8))
latitudes, longitudes = np.array(dp).T
ax = plt.axes(projection=PlateCarree())
ax.scatter(longitudes, latitudes)
ax.set_global()
ax.coastlines()
ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND)
plt.show()
```

**4. Cluster the locations based on geographic distance**
Using K-Means:

```python
#Elbow method to choose the best k
k_values = range(1,10)
inertia_values = [KMeans(k).fit(dp).inertia_ for k in k_values]

#plotting
plt.plot(k_values, inertia_values)
plt.xlabel('K')
plt.ylabel('inertia')
plt.show()
```

```python
#get the clusters for k=3
model = KMeans(n_clusters=3)
clusters = model.fit_predict(dp)
```

```python
def plot_clusters(clusters, longitudes, latitudes):
    plt.figure(figsize=(12, 10))
    ax = plt.axes(projection=PlateCarree())
    ax.coastlines()

    # Identify noise points (typically labeled -1)
    noise_mask = (clusters == -1)
    cluster_mask = ~noise_mask

    # Plot clusters normally
    scatter = ax.scatter(
        longitudes[cluster_mask],
        latitudes[cluster_mask],
        c=clusters[cluster_mask],
        cmap='tab10',
        s=50,
        label='Clusters'
    )

    # Plot noise/outliers with 'x' markers
    ax.scatter(
        longitudes[noise_mask],
        latitudes[noise_mask],
        c='k',
        marker='x',
        s=60,
        label='Noise/Outliers'
    )

    ax.set_global()
    plt.legend()
    plt.title("Clustered Points with Noise Marked as 'X'")
    plt.show()
```

```python
plot_clusters(clusters,longitudes,latitudes)
```

**Clustering using DBSCAN**

```python
#convert longitude,latitude from degree to radian
dp_rad = np.radians(dp)

#DBSCAN clustering
R = 6371.0088  # Earth radius in km
eps_km = 1000
eps_rad = eps_km / R

db_model = DBSCAN(eps=eps_rad, min_samples=3, metric='haversine')
db_clusters = db_model.fit_predict(dp_rad)
```

```python
#plotting
plot_clusters(db_clusters,longitudes,latitudes)
```

```python
for eps_km in [50, 100, 200, 500, 1000]:
    eps_rad = eps_km / R
    db = DBSCAN(eps=eps_rad, min_samples=3, metric='haversine')
    labels = db.fit_predict(dp_rad)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    print(f"eps={eps_km} km → {n_clusters} clusters, noise={(labels==-1).sum()}")
```
